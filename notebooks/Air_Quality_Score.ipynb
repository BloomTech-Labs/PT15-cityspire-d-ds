```
# get the city and state names
def get_state_name(city):
  state = city.split(',')[1]
  state = state.strip()
  return state

def get_city_name(city):
  name = city.split(',')[0]
  name = name.split()
  result = name[0]
  for word in name[1:]:
    if word[0].isupper():
      result += ' '
      result += word
  return result

aq['City'] = aq['Core Based Statistical Area (CBSA)'].apply(get_city_name)
aq['State'] = aq['Core Based Statistical Area (CBSA)'].apply(get_state_name)
aq['city_code'] = aq['City'].apply(lambda x: x.replace(' ', '_'))
```

```
# check for null values
aq.isnull().sum()
```

```
# based on information I could gather, these two columns play the most factor
# in determining air quality, and they also happen to be the two columns with the
# least amount of NaN values so they are the two that I chose to help create
# the air quality score
aq = aq[['City', 'State', 'city_code', 'PM2.5         Wtd AM (µg/m3) ', 'O3            8-hr (ppm)']]
aq = aq.rename({'PM2.5         Wtd AM (µg/m3) ': 'PM2.5', 'O3            8-hr (ppm)': 'O3'}, axis=1)
aq.head()
```

```
# I scaled the data so the two columns could be combined and 
# also used SimpleImputer to get rid of the NaN values
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

imp = SimpleImputer()
aq['O3'] = imp.fit_transform(aq[['O3']])

scl1 = StandardScaler(with_mean=False)
aq['PM2.5 Scaled'] = scl1.fit_transform(aq[['PM2.5']])

scl2 = StandardScaler(with_mean=False)
aq['O3 Scaled'] = scl2.fit_transform(aq[['O3']])

aq.describe()
```

```
aq['Combined Total'] = aq['PM2.5 Scaled'] + aq['O3 Scaled']
```

```

buckets = aq['Combined Total'].quantile([0, .2, .4, .6, .8, 1]).values
buckets
```

```
# another function to return a 1-5 score for air quality based on the above 
# quantiles of the data set
def get_aq_score(combined_aq):
    buckets = [ 7.08867508, 10.12566503, 10.89508254, 11.95768491, 12.90177913,
                16.86715543]
    if buckets[0] <= combined_aq <= buckets[1]:
        score = 5
    elif combined_aq <= buckets[2]:
        score = 4
    elif combined_aq <= buckets[3]:
        score = 3
    elif combined_aq <= buckets[4]:
        score = 2
    elif combined_aq <= buckets[5]:
        score = 1
    return score
```

```
# look at the distribution of scores
x = pd.Series([get_aq_score(total) for total in aq['Combined Total']])
x.value_counts()
```
